```python
df.drop(df.columns[[12]], axis=1, inplace=True)
```

```python
dfTemp = df["Date Recorded"].str.split('/', expand=True)
print (dfTemp)
```

```python
dfTemp.drop([0,1])
print (dfTemp)
```

```python
df= pd.concat ([df,dfTemp], axis=1)
print (df)
```

```python
df.rename({2:"Sale-year"}, axis=1, inplace=True)
df.columns
```

```python
df.info()
```

```python
df['amount']=df['Sale Amount']/1000000
df.head()
```

```python
df.info()
```

```python
df.reset_index()
```

```python
dfing = df.groupby(['Sale-year']).sum('amount')
dfing = dfing.iloc[:, :14]
dfing = dfing.sort_values('amount' ,ascending=False)
dfing
```

```python
dfing.info()
```

```python
mpl.rcParams['axes.spines.bottom']=True
mpl.rcParams['axes.spines.left']=False
mpl.rcParams['axes.spines.right']=False
mpl.rcParams['axes.spines.top']=False
```

```python
fig, ax = plt.subplots(figsize = (15,6.9), dpi=1000)

# Datos a graficar
ingr = dfing.index.tolist()
y_pos = np.arange(len(ingr))

 
ingr2 = dfing['amount'].tolist()
 
# Arreglo de colores
barColors = ['#ebd07f','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5',
             '#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5',
             '#e5e5e5','#e5e5e5','#eb7f7f']
 
# Generacion del grafico
ax.barh(y_pos, ingr2, align='center', color = barColors)
ax.set_yticks(y_pos, labels= ingr)
ax.invert_yaxis()  


#-----------------------------------------------------------------------------------------------------------------------------#
#-----------------------------CONFIGURACION DE LOS TITULOS, SUBTITULOS, EJES DEL GRAFICO--------------------------------------#
#-----------------------------------------------------------------------------------------------------------------------------#
#formato del titulo
 
title = "Cual es el Ingreso anual de las propiedades vendidas?"
plt.suptitle(title,
            horizontalalignment = 'left',
            x = 0.001,
            y = 0.87,
            transform = fig.transFigure,
            fontsize = 20,
            fontweight = "bold",
            fontfamily = "serif")
 
#formato del subtitulo
stitle = "Ingreso de cada año por ventas"
ax.set_title(stitle,
            horizontalalignment = 'left',
            x = 0.001,
            y = 0.80,
            transform = fig.transFigure,
            fontsize = 11,
            fontweight = "regular",
            fontfamily = "serif")
 
#formato del eje X
etiqueta = "Cantidad de USD en Millones"
ax.set_xlabel(etiqueta,
             ha = 'left',
             fontsize = 10,
             fontweight = "regular",
             fontfamily = "serif")

#Configuracion de coordenadas de la etique del eje x
ax.xaxis.set_label_coords(x=0, y= -0.05, transform= ax.transAxes)
 
#configuracion del tamano
ax.tick_params(axis='both', 
               labelsize = 10)

#Congiruacion de la ubicacion del grafico
plt.subplots_adjust(top=0.80)
 
 
plt.show()
```

<h2 Style="color:#2e4347">Que tipo de propiedad tiene mayor facilidad de venta? </h2>

```python
dfres = df.groupby(['Property Type']).count()
dfres = dfres.iloc[:, :8]
dfres = dfres.sort_values('Town', ascending=False)
dfres
```

```python
dfres = dfres.iloc[0:6, :]
dfres
```

```python
mpl.rcParams['axes.spines.bottom']=True
mpl.rcParams['axes.spines.left']=False
mpl.rcParams['axes.spines.right']=False
mpl.rcParams['axes.spines.top']=False
```

```python
fig, ax = plt.subplots(figsize = (15,6.9), dpi=1000)

# Datos a graficar
rest = dfres.index.tolist()
y_pos = np.arange(len(rest))

 
cant = dfres['Town'].tolist()
 
# Arreglo de colores
barColors = ['#ebd07f','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#eb7f7f']
 
# Generacion del grafico
ax.barh(y_pos, cant, align='center', color = barColors)
ax.set_yticks(y_pos, labels= rest)
ax.invert_yaxis()  


#-----------------------------------------------------------------------------------------------------------------------------#
#-----------------------------CONFIGURACION DE LOS TITULOS, SUBTITULOS, EJES DEL GRAFICO--------------------------------------#
#-----------------------------------------------------------------------------------------------------------------------------#
#formato del titulo
 
title = "Que tipo de propiedad tiene mayor facilidad de venta?"
plt.suptitle(title,
            horizontalalignment = 'left',
            x = 0.001,
            y = 0.87,
            transform = fig.transFigure,
            fontsize = 20,
            fontweight = "bold",
            fontfamily = "serif")
 
#formato del subtitulo
stitle = "Tipos de propiedades"
ax.set_title(stitle,
            horizontalalignment = 'left',
            x = 0.001,
            y = 0.80,
            transform = fig.transFigure,
            fontsize = 11,
            fontweight = "regular",
            fontfamily = "serif")
 
#formato del eje X
etiqueta = "Cantidad"
ax.set_xlabel(etiqueta,
             ha = 'left',
             fontsize = 10,
             fontweight = "regular",
             fontfamily = "serif")

#Configuracion de coordenadas de la etique del eje x
ax.xaxis.set_label_coords(x=0, y= -0.05, transform= ax.transAxes)
 
#configuracion del tamano
ax.tick_params(axis='both', 
               labelsize = 10)

#Congiruacion de la ubicacion del grafico
plt.subplots_adjust(top=0.80)
 
 
plt.show()
```

<h2 Style="color:#2e4347">Que tipo de propiedad tiene mas facilidad de venta?</h2>

<p> Despues de efectuar el analisis del Data Set se determino que las propiedades del tipo residencial tienen un predominancia en la la cantidad de ventas efectuadas por la empresa, por lo cual la recomendacion es dirigir la mayor cantidad de esfuerzos a las ventas de este tipo de propiedades.
