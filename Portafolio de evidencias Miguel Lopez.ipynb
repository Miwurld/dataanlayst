<h2 Style="color:#2e4347">2. manejo de datos faltantes</h2>

<p> se realiza la carga de los set de datos para iniciar con la revision de la estructura general y de esta forma lograr crear un data frame donde se consoliden las diferentes fuentes identificadas.</p>

<ol>
    <li>identificar como se muestran los valores faltantes</li>
    <li>Contar la cantidad de valores faltante en el set de datos</li>

<h3 Style="color: #bac9a9"> 2.1 Identificar los valores faltantes</h3>


```python
dfAutos.replace('?', np.nan, inplace = True)
dfAutos.head(6)                
```

<h3 Style="color: #bac9a9"> 2.2 Contar los valores faltantes</h3>

```python
dfTemp = dfAutos.isnull()
dfTemp.head(6)
```

```python
for clumn in dfTemp.columns.values.tolist():
    print(clumn)
    print(dfTemp[clumn].value_counts())
    print("")
```

<h3 Style="color: #bac9a9"> 2.3 Imputar valores numericos</h3>

```python
avNormalized = dfAutos['normalized'].astype("float").mean(axis=0)
avPeak = dfAutos['peak'].astype("float").mean(axis=0)
avHorsepower = dfAutos['horsepower'].astype("float").mean(axis=0)
avBore = dfAutos['bore'].astype("float").mean(axis=0)
avStroke= dfAutos['stroke'].astype("float").mean(axis=0)                 
```

```python
dfAutos['normalized'].replace(np.nan, avNormalized, inplace = True)
dfAutos['peak'].replace(np.nan, avPeak, inplace = True)
dfAutos['horsepower'].replace(np.nan, avHorsepower, inplace = True)
dfAutos['bore'].replace(np.nan, avBore, inplace = True)
dfAutos['stroke'].replace(np.nan, avStroke, inplace = True)
```

```python
dfAutos.head(6)
```

<h3 Style="color: #bac9a9"> 2.4 Imputar valores categoricos/ caracter</h3>

```python
dfAutos['doors'].value_counts().idxmax()
```

```python
dfAutos['doors'].replace(np.nan, "four", inplace = True)
```

<h3 Style="color: #bac9a9"> 5.1 Verificar el dominio de la variables</h3>

```python
dfAutos[["normalized"]] = dfAutos[["normalized"]].astype("int")
dfAutos[["price", "horsepower", "peak"]] = dfAutos[["price", "horsepower", "peak"]].astype("int")
dfAutos[["bore","stroke"]] = dfAutos[["bore","stroke"]].astype("float")
```

```python
dfAutos.describe()
```

<h3 Style="color: #bac9a9"> 5.2 Convertir valores numericos en funcion del medio (Estandarizar)</h3>

```python
dfAutos['City/100km(Calc)']= 235.21/dfAutos['City']
dfAutos['HighwayL/100km(Calc)']= 235.21/dfAutos['Highway']
dfAutos.head(5)
```

<h3 Style="color: #bac9a9"> 5.3 Normalizar valores entre 0 y 1</h3>

```python
dfTemp =dfAutos[['lenght','height']]
dfTemp.head(4)
```

```python
dfAutos['lenght(Norm)']= dfAutos['lenght']/dfAutos['lenght'].max()
dfAutos['height(Norm)']= dfAutos['height']/dfAutos['height'].max()
dfAutos.head(4)
```

<h2 Style="color:#2e4347">Cual es la cantidad de datos segun el fabricante?</h2>

```python
gf = dfAutos.groupby(['maker']).count()
gf = gf.iloc[:,:1]
gf = gf.sort_values('symboling', ascending=True)
gf
```

```python
fig, ax = plt.subplots(1,1, figsize = (15,6.9), dpi=100)
 
# Datos a graficar
maker = gf.index.tolist()
y_pos = np.arange(len(maker))
 
cant = gf['symboling'].tolist()
 
# Arreglo de colores
barColors = ['#ebd07f','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5',
             '#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5',
             '#e5e5e5','#e5e5e5','#e5e5e5','#eb7f7f']
 
# Generacion del grafico
ax.barh(y_pos, cant, align='center', color = barColors)
ax.set_yticks(y_pos, labels=maker)
ax.invert_yaxis()  # labels read top-to-bottom
#-----------------------------------------------------------------------------------------------------------------------------#
#-----------------------------CONFIGURACION DE LOS TITULOS, SUBTITULOS, EJES DEL GRAFICO--------------------------------------#
#-----------------------------------------------------------------------------------------------------------------------------#
#formato del titulo
 
title = "Grafico 1: Vehiculos en Inventario"
plt.suptitle(title,
            horizontalalignment = 'left',
            x = 0.001,
            y = 0.87,
            transform = fig.transFigure,
            fontsize = 20,
            fontweight = "bold",
            fontfamily = "serif")
 
#formato del subtitulo
stitle = "¿Cuál es la cantidad de datos según fabricante?"
ax.set_title(stitle,
            horizontalalignment = 'left',
            x = 0.001,
            y = 0.80,
            transform = fig.transFigure,
            fontsize = 11,
            fontweight = "regular",
            fontfamily = "serif")
 
#formato del eje X
etiqueta = "Cantidad de vehiculos"
ax.set_xlabel(etiqueta,
             ha = 'left',
             fontsize = 10,
             fontweight = "regular",
             fontfamily = "serif")
 
#Configuracion de coordenadas de la etique del eje x
ax.xaxis.set_label_coords(x=0, y= -0.05, transform= ax.transAxes)
 
#configuracion del tamano
ax.tick_params(axis='both', 
               labelsize = 10)
 
#Congiruacion de la ubicacion del grafico
plt.subplots_adjust(top=0.80)
 
 
plt.show()
```
